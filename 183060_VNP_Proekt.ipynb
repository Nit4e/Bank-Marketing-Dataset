**<font color = "#0F72A0"> Можеме да забележиме дека има големи разлики помеѓу вредностите кај некои од нумеричките колони. За ова ќе треба да направиме дополнителни обработки. </font>**</br>
**<font color = "#0F72A0"> Прво, според вредностите од хистограмите, ќе определиме кои вредности отскокнуваат од останатите. </font>**

```python
a = len (dataset[dataset['duration'] > 3000] ) / len(dataset) * 100
b = len (dataset[dataset['campaign'] > 35] ) / len(dataset) * 100
c = len (dataset[dataset['pdays'] > 800] ) / len(dataset) * 100
d = len (dataset[dataset['previous'] > 35] ) / len(dataset) * 100
print("Vrednosti od duration nad 3000: ", "{:.2f}".format(a),
      "%\nVrednosti od campaign nad 35: ", "{:.2f}".format(b),
      "%\nVrednosti od pdays nad 800: ", "{:.2f}".format(c),
      "%\nVrednosti od previous nad 35: ", "{:.2f}".format(d), "%")
```

**<font color = "#0F72A0"> Бидејќи вредностите поголеми од 3000, 35, 800 и 35 кај 'duration', 'campaign', 'pdays' и 'previous' соодветно, се невообичаени, односно процентот на вакви одговори е многу мал, тие ќе ги заменам со просекот од вредностите во секоја од колоните посебно. Ова го правам со цел подоцна овие вредности да не влијаат врз резултатите бидејќи иако се во мал број, сепак можат да влијаат врз крајниот исход. </font>**


```python
dataset[dataset['duration'] > 3000] = round(dataset['duration'].mean())
dataset[dataset['campaign'] > 35] = round(dataset['campaign'].mean())
dataset[dataset['pdays'] > 800] = round(dataset['pdays'].mean())
dataset[dataset['previous'] > 35] = round(dataset['previous'].mean())
```

### <font color = '#900C3F'> **Keras** </font>

```python
X = data.drop(['deposit'], axis=1)
Y = data.deposit.values

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = True)
```

```python
X = np.asarray(X)
Y = np.asarray(Y)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = True)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense

modelNN = Sequential()

modelNN.add(Dense(650, input_dim = 15, activation="sigmoid"))
modelNN.add(Dense(50, activation="sigmoid"))
modelNN.add(Dense(1, activation="sigmoid"))
```

```python
modelNN.compile(loss="mse", optimizer="adam")
```

```python
historyNN = modelNN.fit(X_train, Y_train, epochs=150, batch_size=185, verbose=1, validation_split=0.2)
```

```python
plt.plot(historyNN.history['loss'])
plt.plot(historyNN.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc = 'upper left')
plt.figure(figsize=(5, 5))
plt.show()
```
